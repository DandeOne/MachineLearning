## Again using the same Position - Salary dataset

Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,-1].values
```

Fitting the model

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 42)
regressor.fit(X,y)
```

Predicting

```python
prediction = regressor.predict([[6.5]])
prediction
```

Visualization

```python
plt.scatter(X,y,color='red')
plt.plot(X,regressor.predict(X),color='blue')
plt.title('Salary predction based on position level')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

A little trap because of how tree algorithm works. It is non-continous model but here the plot is combining the prediction values

Another variant of visualization

```python
X_more_values = np.arange(min(X),max(X),0.01)
X_more_values = X_more_values.reshape(len(X_more_values),1)
plt.scatter(X,y,color='red')
plt.plot(X_more_values,regressor.predict(X_more_values),color='blue')
plt.title('Salary predction based on position level')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
